## 3.1 Mean Squared Error (MSE )

```python
def mse_cal(y,h):
  return np.mean((y-h)**2)
```

```python
np.mean((y-h)**2)
```

```python
mse_cal(y,h)
```

### Loss function
idea Loss function : ต้องการตัวชี้วัดเป็นตัวเลขค่าเดียวที่บอกว่าโมเดลของเราทำงานได้ดีแค่ไหน โดยเปรียบเทียบ output ของ โมเดล กับข้อมูล output ตัวอย่าง y ค่าทั้งสองอย่าง ณ input x เดียวกัน

```python
def Loss_func(w):
  h = w*x
  return mse_cal(y,h)
```

```python
# ค่าของ loss function ที่ได้เมื่อกำหนด w = 0
Loss_func(0)
```

### วิธีที่ 1 หาค่า optimal parameter โดยการสุ่มค่า w
การหาค่า optimal parameter ที่ง่ายที่สุด คือ การลองสุ่มค่า w ขึ้นมาจำนวนมาก แล้วทำลูปให้ทดลองทุกค่า w ที่สุ่มมา จากนั้นดูว่าค่า w ค่าไหน ที่ทำให้โมเดลมีค่า error ต่ำที่สุด

*วิธีการนี้จะใช้ได้เฉพาะกรณีที่มี* parameter จำนวนน้อย ๆ เท่านั้น

```python
# สร้าง w หลายๆตัวไว้สำหรับทดลองเพื่อหาค่า w ที่ดีที่สุด
wList = np.arange(-200,240)/10
```

```python
wList
```

```python
# วิธีหาค่า parameter อย่างง่าย

# สร้าง list สำหรับบันทึก error ของแต่ละ W
mseList = []

# random wList
wList = np.arange(-200,240)/10

# loop w ทุกตัวใน wList
for w in wList:
  mse = Loss_func(w)
  mseList.append(mse)

# plot Loss function
py.scatter(wList, mseList)
```

### วิธีที่ 2 หาค่า parameter ที่ดีที่สุดโดยใช้ Gradient descent
ในความเป็นจริง โมเดลจะมีความซับซ้อนมาก จำนวน parameter ก็มีจำนวนมากตามไปด้วย จึงไม่สามารถใช้วิธีการสุ่มได้ ซึ่งในปัจจุบันโมเดลส่วนใหญ่นิยมใช้ Gradient descent ในการหาค่า optimal parameter แทน

Loss function equation

```python
# dj by dw
np.mean((h - y) * x)
```

```python
# initialize weight
w = -10000

mse = Loss_func(w)

# learning rate
alpha = 0.0001

# number of iteration
iter = 20

wList = [w]
mseList = [mse]
```

```python
mse
```

ทดลองปรับค่า w โดยใช้ gradient descent

```python
# initial w value = -10000
w
```

```python
for i in range(iter):
  h = w * x
  dj_by_dw = np.mean((h - y) * x)
  w = w - alpha * dj_by_dw
```

```python
# updated w สำหรับการทำซ้ำ 20 ครั้ง
w
```

พลอตกราฟดูการขยับเส้น h ตาม w ที่เปลี่ยนแปลงไป

```python
# re initialize w = -10000
w = -10000
```

```python
for i in range(iter):
  h = w * x
  dj_by_dw = np.mean((h - y) * x)
  w = w - alpha * dj_by_dw

  fig, (ax1, ax2) = py.subplots(1, 2)
  mseList.append(mse)
  wList.append(w)
  ax1.scatter(x, y)
  ax1.plot(x,x*w,c='r')
  ax2.scatter(wList, mseList)
  py.show()
```

สร้าง function train_model เพื่อทดลองค่า hyperparameter ค่าต่าง ๆ

```python
def train_model(iter=20, w=-10000, alpha=0.0001):
  mse = Loss_func(w)
  wList = [w]
  mseList = [mse]
  for i in range(iter):
    h = w * x
    dj_by_dw = np.mean((h - y) * x)
    w = w - alpha * dj_by_dw
    fig, (ax1, ax2) = py.subplots(1, 2)
    mseList.append(mse)
    wList.append(w)
    ax1.scatter(x, y)
    ax1.plot(x,x*w,c='r')
    ax2.scatter(wList, mseList)
    py.show()
  return wList, mseList
```

```python
# ลอง train ให้นานขึ้นจาก 20 -> 50
w_list, loss_list = train_model(iter=50)
```
